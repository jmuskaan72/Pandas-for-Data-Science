## Making changes to the data

```python
#df.head(15)
#ADD TOTAL AS A PARAMETER in Dataframe CONTAINING ALL THE FEATURES IN A WAY TO DECIDE WHICH POKEMON IS THE BEST!
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

#OR we can also use dropping columns and use iloc method for total. 
#Also, axis 1 takes horizontally and axis 0 takes vertially values
df  = df.drop(columns='Total')
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

#rearranging columns
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(15)
```

## Regex filtering : Textual patterns

```python
#1. LIST ALL THE POKEMONS THAT DOESN'T CONTAIN 'Mega' NAME IN IT (drop them using (~)).
df.loc[~df['Name'].str.contains('Mega')] 

import re
#2. It will check if Type 1 has fire or grass pokemons and re.I is ignoring the captilization needed for F in Fire or in Grass.
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]

#3. To get all the pokemons names starting with 'Pi'. 
# ^ will prevent to show any letter having 'pi' in between e.g.'Caterpie' and * means Zero or more values present.
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```
